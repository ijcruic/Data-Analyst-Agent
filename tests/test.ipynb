```python
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain_mcp_adapters.resources import load_mcp_resources
from mcp.shared.exceptions import McpError
```

```python
  MCP_ENDPOINT = "http://localhost:4242/mcp"
  URI = "mcp://tmp/mcp_outputs/0f2d3379-2fd5-4f6a-8dd2-5b39a5db09ab/total_spend_by_contract_type.png"  # replace with the exact URI

  client = MultiServerMCPClient(
      {
          "mcp_server": {
              "transport": "streamable_http",
              "url": MCP_ENDPOINT,
          }
      }
  )

  session_manager = client.session("mcp_server")
  session = await session_manager.__aenter__()
```

```python
blobs = await client.get_resources("mcp_server")  
```

```python
  try:
      try:
          blobs = await load_mcp_resources(session, uris=[URI])
      except McpError as exc:
          print(f"MCP reported an error: {exc}")
      else:
          if not blobs:
              print("No resource blobs returned.")
          else:
              blob = blobs[0]
              print(f"Resource URI: {URI}")
              print(f"  mime_type: {blob.mime_type!r}")
              if blob.text is not None:
                  print(f"  text payload (first 200 chars): {blob.text[:200]!r}")
              else:
                  data = blob.data or b""
                  if isinstance(data, str):
                      data = data.encode("utf-8")
                  print(f"  binary payload length: {len(data)} byte(s)")
```
